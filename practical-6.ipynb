#Graphical Representation

Bar Graph

```python
plt.subplot(221)
df['gender'].value_counts().plot(kind='bar',title='Gender of students',figsize=(16,9))
plt.xticks(rotation=0)

plt.subplot(222)
df['race/ethnicity'].value_counts().plot(kind='bar',title='Race/Ethnicity of students',figsize=(16,9))
plt.xticks(rotation=45)

plt.subplot(223)
df['lunch'].value_counts().plot(kind='bar',title='Lunch Type of students',figsize=(16,9))
plt.xticks(rotation=0)

plt.subplot(224)
df['test preparation course'].value_counts().plot(kind='bar',title='Test Preparation Course of students',figsize=(16,9))
plt.xticks(rotation=0)
```

```python
df.boxplot()
```

```python
sns.displot(df['math score'])
```

```python
sns.displot(df['math score'],bins=5)
```

```python
corr=df.iloc[:,5:].corr()
sns.heatmap(corr,annot=True)
plt.yticks(rotation=0)
plt.show()
```

```python
sns.relplot(x='math score',y='writing score',data=df,hue='gender')
```

```python
df.groupby('test preparation course')[['math score','reading score','writing score']].mean().T.plot(kind='barh',figsize=(5,5))
```
