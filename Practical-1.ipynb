#Set

```python
x={"apple","banana","cherry"}
y={"google","microsoft","apple"}

z=x.union(y)
print(z)
```

```python
z=x.intersection(y)
print(z)
```

```python
z=x.difference(y)
print(z)
```

```python
z=y.difference(x)
print(z)
```

```python
z=x.isdisjoint(y)
print(z)
```

```python
z=x.issubset(y)
print(z)
```
