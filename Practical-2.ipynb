#Loops

For Loop

```python
got_house=['stark','arryn','barayheon']
```

```python
for house in got_house:
  print(f"House {house}")
```

```python
for number,house in enumerate(got_house):
  print(f"The House no of {house} is {number+1}")
```

```python
lofn=[1,22,333,4444,55555]
```

```python
sqr_lofn=[]
for number in lofn:
  s=number**2
  sqr_lofn.append(s)
print(sqr_lofn)
```

#Functions & Methods

```python
def extracr_first_letter(sentence):
  words_in_sentence = sentence.split(' ')
  print(words_in_sentence)

  first_letters=[word[0] for word in words_in_sentence]
  return(first_letters)
```

```python
starwars_quote='May the force be with you'
first_letters=extracr_first_letter(starwars_quote)
print(first_letters)
```
